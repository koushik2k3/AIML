```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./advertising.csv')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
for i in df.columns:
    print(i)
    print(df[i].unique())
```

```python
df.head()
```

```python
print(len(df['Ad Topic Line'].unique()),df.shape)
print(len(df['City'].unique()),df.shape)
print(len(df['Country'].unique()),df.shape)
```

```python
df['Timestamp']=pd.to_datetime(df['Timestamp'])
```

```python
df.drop('Timestamp',axis=1,inplace=True)
```

```python
enc = LabelEncoder()
scaler = StandardScaler()
```

```python
df['Country'] = enc.fit_transform(df['Country'])
df['Ad Topic Line'] = enc.fit_transform(df['Ad Topic Line'])
df['City'] = enc.fit_transform(df['City'])
```

```python
df
```

```python
X = df.drop('Clicked on Ad',axis=1)
Y = df['Clicked on Ad']
```

```python
scaled_x = scaler.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
lr = LogisticRegression()
nb = GaussianNB()
knn = KNeighborsClassifier(n_neighbors=5)
dt = DecisionTreeClassifier()
```

```python
lr.fit(x_train,y_train)
nb.fit(x_train,y_train)
knn.fit(x_train,y_train)
dt.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
nb_pred = nb.predict(x_test)
knn_pred = knn.predict(x_test)
dt_pred = dt.predict(x_test)
```

```python
lr.fit(x_train_scaled,y_train)
nb.fit(x_train_scaled,y_train)
knn.fit(x_train_scaled,y_train)
dt.fit(x_train_scaled,y_train)
```

```python
lr_pred = lr.predict(x_test_scaled)
nb_pred = nb.predict(x_test_scaled)
knn_pred = knn.predict(x_test_scaled)
dt_pred = dt.predict(x_test_scaled)
```

```python
results_df = pd.DataFrame({'prediction': lr_pred, 'test_label': y_test})
results_df # Display the first few rows
```

```python
results_df = pd.DataFrame({'prediction': nb_pred, 'test_label': y_test})
results_df # Display the first few rows
```

```python
results_df = pd.DataFrame({'prediction': knn_pred, 'test_label': y_test})
results_df # Display the first few rows
```

```python
results_df = pd.DataFrame({'prediction': dt_pred, 'test_label': y_test})
results_df # Display the first few rows
```

```python
# Logistic Regression
report = classification_report(y_test, lr_pred)
print(report)
```

```python
# Naive Bayes
report = classification_report(y_test, nb_pred)
print(report)
```

```python
# KNN
report = classification_report(y_test, knn_pred)
print(report)
```

```python
# Decision Tree
report = classification_report(y_test, dt_pred)
print(report)
```
