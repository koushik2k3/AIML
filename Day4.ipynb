# Feature Scaling

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("BreastCancer.csv")
```

```python
df.head(5)
```

```python
x=[1000,2000,3000,5000,9000]
y=[9000,5000,3000,2000,1000]
```

```python
avg = sum(x)/len(x)
```

```python
sum_x = 0
for i in x:
    temp = (i-avg) ** 2
    sum_x+= temp
sd = (sum_x/(len(x)-1)) ** 0.5
```

```python
sd
```

```python
import numpy as np
```

```python
np.std(x)
```

```python
z_score = []
for i in x:
    z_score+= [(i-avg)/sd]
```

```python
z_score
```

```python
plt.scatter(x,y,label="stars",color="blue",marker="*",)
plt.legend()
plt.show()
```

```python
ax = sns.scatterplot(
                     x=x,
                     y=y
                     )
ax.set_title("Normalized")
plt.show()
```

```python
ax = sns.scatterplot(
                     x=z_score,
                     y=y
                     )
ax.set_title("Normalized")
plt.show()
```
