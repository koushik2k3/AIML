# quantiles


```python
# Q1=np.percentile(ds['area_mean'],25, interpolation='midpoint')

# Q3=np.percentile(ds['area_mean'],75,interpolation='midpoint')

# IQR= Q3 - Q1

# print("Old Shape: ",ds.shape)

# #upper bound
# upper=np.where(ds['area_mean']>=(Q3+1.5*IQR))

# #lower bound
# lower=np.where(ds['area_mean']<=(Q1-1.5*IQR))


# #removing outliers

# ds.drop(upper[0], inplace=True)
# ds.drop(lower[0], inplace=True)
# print("New Shape: ",ds.shape)
```

```python
#method 2

Q1=ds.area_mean.quantile(0.25)
Q1
Q3=ds.area_mean.quantile(0.75)
Q3
IQR=Q3-Q1
IQR

Lower=Q1-1.5*IQR
Upper=Q3+1.5*IQR
```

```python
IQR
```

```python
ds[(ds.area_mean<Lower)|(ds.area_mean>Upper)]
```

```python
ds_no_out=ds[(ds.area_mean>Lower)&(ds.area_mean<Upper)]  #removing outliers and storing them in a new df
```

```python
ds_no_out.head(2)
```

```python
sns.boxplot(x=ds_no_out['area_mean'])   #doesnt remove all the outliers but it reduces the number of outliers
```

```python
#method 3
#replacing values which are more than upper as upper, and the ones less than lower as lower value
#capping method

new_df=ds.copy()

new_df.area_mean=np.where(
    new_df.area_mean>Upper,
    Upper,
    np.where(
       new_df.area_mean<Lower,
       Lower,
       new_df.area_mean
    )
)
```

```python
sns.boxplot(x=new_df['area_mean'])
```

# mean, stdev, normalize practice


```python
x=[1000,2000,3000,5000,9000]

```

```python
n=len(x)
mean_ans=0
ans=0
j=0
```

```python
#mean

for i in x:
    j=j+i
    mean=j/n
mean
```

```python
#STD

import math
x_sum = 0
for i in range(len(x)):
   
    x_sum+=(x[i]- mean)**2
    
std = math.sqrt(x_sum/(len(x) - 1))

std
```

```python
#Normalize

z_norm=[]
count=0
for i in x:
    z=(i-mean)/std
    print(z)
    z_norm.append(z)
    
```
